# Find and count missing values: 

```python
# Creating a loop to go through columns and calculate the missing values
for col in stats.columns:
    pct_missing = stats[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

# Groupby/Aggregating 

```python
#Group stats by alignment - mean()
grouped_average = stats.groupby(['Alignment']).mean().sort_values('Total', ascending=False)
print(grouped_average)
```

```python
#Group info by Race - sum()
grouped_race = info.groupby(['Race']).sum()
print(grouped_race.head())
```

```python
#Group info by Gender - count
grouped_gender = info.groupby(['Gender']).count()
print(grouped_gender.head())
```

```python
eyes = info.groupby('EyeColor')[['Name']].count().sort_values('Name', ascending=False).head(5).reset_index()
hair = info.groupby('HairColor')[['Name']].count().sort_values('Name', ascending=False).head(5).reset_index()
eyes = eyes.rename(columns = {'variable': 'Name', 'value': 'Eye_Count'}, inplace = False)
hair = hair.rename(columns = {'variable': 'Name', 'value': 'Hair_Count'}, inplace = False)
corr = pd.concat([eyes, hair], axis=1)
corr = corr.corr()
print(corr)
```

# Loops/iterrows 

```python
#Loop to see who has the longest name
for lab, row in info.iterrows() :
    info.loc[lab, "name_length"] = len(row["Name"])

info["name_length"] = info["Name"].apply(len)
print(info.head())
```

# Merge

```python
superheroes = pd.merge(info, stats, on="Name", how="inner")
print(superheroes.shape)
print(superheroes.info())
```

```python
# Exploring Speed to subset data into categories 
stats.Speed.describe()
```

```python
# Creating categorical column for Speed
def create_cat(i):
    if i >= 0 and i <=35:
        return 'Quick'
    if i >= 35 and i <=70:
        return 'Fast'
    if i >= 70 and i <=100:
        return 'SuperSpeed'
stats['Speed'] = stats['Speed'].apply(create_cat)
```

```python
#Women villains 
W_villain = superheroes.loc[(superheroes['Gender'] == 'Female') & (superheroes['Alignment_y'] == 'bad')]
W_villain = W_villain['Name'].value_counts().head(5)
W_villain.plot(kind = 'pie');
```

```python
#Histogram
info["Gender"].fillna("No Gender", inplace = True)
info["Alignment"].fillna("No Alignment", inplace = True)
```

```python
sns.displot(info, x='Gender', hue='Alignment', palette=("PuBuGn"))
plt.title('Gender Distribution through Characters Alignment');
```

```python
sns.displot(stats, x='Speed', hue='Alignment', palette=("PuBuGn"));
plt.title('How fast are superheroes?');
```

```python
#Pie chart
info["Race"].fillna("No Race", inplace = True)
races_pie = info['Race'].value_counts().head(5)
plt.style.use('seaborn-colorblind')
races_pie.plot(kind = 'pie') 
plt.title('Races in the Marvel World')
```

```python
#Bar Chart
powerful_heroes = superheroes.groupby('Name')[['Total']].sum().sort_values('Total', ascending=False).head(5).reset_index()
sns.barplot(x="Name", y="Total", data=powerful_heroes, palette=("PuBuGn"))
plt.title('The Most Powerful Heroes');
```

```python
#Bar Chart
publishers = superheroes.groupby('Publisher')[['Name']].count().sort_values('Name', ascending=False).head(5).reset_index()
sns.barplot(x="Publisher", y="Name", data=publishers, palette=("PuBuGn"))
plt.title('Publishers behind the Marvel Heroes');
```
